# Read sounds

```python
%matplotlib inline
import matplotlib.pyplot as plt
from scipy.io import wavfile as wav
from glob import glob
```

```python
mySound_list = [wav.read(hello, 'r') for hello in glob("sound/*.wav")]
```

# Plot sounds

```python
# the first element is rate, the second is the values :)
for rate, mySound in mySound_list:
    plt.plot(mySound)
    plt.show()
```
